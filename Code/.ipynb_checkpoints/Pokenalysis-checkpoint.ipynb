```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
path = '../resources/pokedex_(Update.04.20).csv'

pokedex = pd.DataFrame(pd.read_csv(path))

pokedex.head()
```

```python
gen_count = pokedex.groupby('generation').size()
gen_count
```

```python
gen_count.plot.bar(title='Pokemon Added By Generation', figsize=(10,8))
plt.ylabel('Number of Pokemon')
```

```python
type_count = pokedex.groupby('type_1').size().sort_values(ascending=False)
type_count
```

```python
type_count.plot.bar(figsize=(12,8), title='Pokemon Primary Types (Gen 1-8)')
plt.ylabel('Number of Pokemon')
```

```python
type2_count = pokedex.groupby('type_2').size().sort_values(ascending=False)
type2_count.plot.bar(figsize=(12,8), title='Pokemon Secondary Types (Gen 1-8)')
plt.ylabel('Number of Pokemon')
```

```python
gen_type = pokedex.groupby(['type_1', 'generation'])['name'].count().unstack('generation').fillna(0)
gen_type
```

```python
gen_type.plot.bar(stacked=True, figsize=(12,10), title='Pokemon Type by Generation')
plt.ylabel('Number of Pokemon')
plt.xlabel('Primary Type')
```

```python
flipped_gen_type = gen_type.T
flipped_gen_type.plot.line(figsize=(12,8))
```

```python
legendary = pokedex.groupby(['is_legendary', 'is_mythical', 'generation'])['name'].count().unstack('generation').fillna(0).T

legendary.plot.bar(stacked=True, figsize=(12,10), title='Legendary/Mythical Pokemon by Generation')
plt.ylabel('Number of Pokemon')
plt.xlabel('Legendary/Mythical Status')
plt.legend(['Sub-legendary', 'Mythical', 'Legendary'])
```

```python
def scatter_plot():
    # A fuction to create scatter plots from Pokedex data
    
    x = (input('Enter a stat: ').lower())
    y = (input(f'Enter a stat to compare {x} to: ').lower())
    
    x_values = pokedex[x].dropna(how='any')
    y_values = pokedex[y].dropna(how='any')
    
    (slope, intercept, r_value, p_value, std_err) = st.linregress(x_values, y_values)
    x_value = np.asarray(x_values, dtype=np.float64)
    regress_values = x_value * slope + intercept

    # Plots scatter plot and regresion
    plt.figure(figsize=(8, 6))
    plt.scatter(x_values, y_values)
    plt.title(f'Pokemon {x.title()} vs. {y.title()}')
    plt.xlabel(x.title())
    plt.ylabel(y.title())
    plt.plot(x_value,regress_values,"r-")
    
    # Annotates graph with equation
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.annotate(line_eq,xy=(min(x_values),min(y_values)),fontsize=15,color="red")
    
    print(f'The r-squared is: {r_value}')
    
    # Print the p-value
    print(f'The p-value is: {p_value}')
```

```python
scatter_plot()
```
